```python
# Seperating Data and Labels

X = diabetes_dataset.drop(columns = 'Outcome', axis = 1) #dropping particular coloumn Outcome
Y = diabetes_dataset['Outcome']                          #storing all labels in Y
```

Data Standardisation

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
standardized_data
```

```python
X = standardized_data                  # Data
Y = diabetes_dataset['Outcome']       # Model
```

```python
X
```

```python
Y
```

> Training and Testing Data Split

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
X.shape , X_train.shape , X_test.shape
```

> Training the Model

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
# Training the Support Vector Machine Classifier

classifier.fit(X_train, Y_train)
```

> Evaluate our Model --> Model Evaluation

* Finding the Accuracy Score >

```python
# accuracy score on the training data

X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy Score of the Training Data : ', training_data_accuracy)
```

```python
accuracy_percentage = training_data_accuracy * 100
print('Accuracy Percentage of the Training Data : ', accuracy_percentage , '%')
```

> Accuracy Score for Test Data

```python
# accuracy score on the training data

X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy Score of the Test Data : ', test_data_accuracy)
```

```python
accuracy_percentage_test = test_data_accuracy * 100
print('Accuracy Percentage of the Test Data : ', accuracy_percentage_test , '%')
```

Making a Predictive System

```python
input_data = (4,110,92,0,0,37.6,0.191,30)

# Change the Input Data into Numpy Arrays
input_data_as_numpy_array = np.asarray(input_data)

#reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)
```

```python
if(prediction[0] == 0):
  print('The Person is Non-Diabetic')
else:
  print('The Person is Diabetic')
```

> The Model has predicted correctly and said the data is non diabetic by predicting the value to be zero [0]

* Hence denoting 'The Person is Non-Diabetic' .
